```python
# List of columns to convert to datetime
time_columns = [
    'SunStart_Time', 'MonStart_Time', 'TuesStart_Time', 'WedStart_Time',
    'ThuStart_Time', 'FriStart_Time', 'SatStart_Time',
    'SunEnd_Time', 'MonEnd_Time', 'TuesEnd_Time', 'WedEnd_Time',
    'ThuEnd_Time', 'FriEnd_Time', 'SatEnd_Time'
]





# Step 1: Convert time columns to 12-hour format, preserving NaT and blank entries
def convert_to_12_hour_format(time_str):
    if pd.isna(time_str) or time_str == 'NaT':
        return time_str  # Preserve NaT or None (keep blanks)
    
    # Convert to datetime and then format to 12-hour
    try:
        # Using %H:%M:%S for the input format as it contains seconds
        return pd.to_datetime(time_str, format='%H:%M:%S').strftime('%I:%M %p')
    except ValueError:
        return time_str  # If conversion fails, return the original string

# Apply the conversion function to each time column
for col in time_columns:
    railworkers_df[col] = railworkers_df[col].apply(convert_to_12_hour_format)

# Check the DataFrame after conversion
print("\nDataFrame After Conversion to 12-hour Format:")
print(railworkers_df[time_columns])




```

```python
print(railworkers_df['SunStart_Time'])
```

```python
#Convert all float64 columns to integers
float_cols = railworkers_df.select_dtypes(include=['float64']).columns

for col in float_cols:
    # Fill NaN values with 0 (or any other appropriate value) before conversion
    railworkers_df[col].fillna(0, inplace=True)
    railworkers_df[col] = railworkers_df[col].astype(int)

# Display the first few rows to confirm changes
print(railworkers_df.head())

# Check data types after conversion
print(railworkers_df.dtypes)
```

```python
# Remove a single column
railworkers_df.drop('Older_Younger_or_Same', axis=1, inplace=True)                
```
