```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Prepare the data
# Define X and y
features = sleep_remote_df.drop(columns=['Satisfaction_with_Remote_Work','Employee_ID'])  # Drop irrelevant columns like Employee_ID
y = sleep_remote_df['Satisfaction_with_Remote_Work']

# Step 2: Encode the target variable if it's categorical
if y.dtype == 'object':
    le = LabelEncoder()
    y = le.fit_transform(y)  # "poor" -> 0, "average" -> 1, "good" -> 2 (example encoding)

# Step 3: Encode categorical features in X
X = pd.get_dummies(features, drop_first=True)

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Train a Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Step 6: Evaluate the model
y_pred = model.predict(X_test)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))

# Step 7: Get feature importances
feature_importances = model.feature_importances_
features_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})
features_df = features_df.sort_values(by='Importance', ascending=False)

# Step 8: Plot feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=features_df.head(10))  # Show top 10 most important features
plt.title('Top 10 Features Contributing to Satisfaction with Remote Work')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.savefig('top10_sleep.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select and scale features
features = sleep_remote_df[['Work_Life_Balance_Rating', 'Satisfaction_with_Remote_Work',
                            'Stress_Level_1', 'Stress_Level_2', 'Sleep_Quality', 
                         'Social_Isolation_Rating',
                            'Mental_Health_Condition', 'Access_to_Mental_Health_Resources',
                            'Physical_Activity_Weekly', 'Company_Support_for_Remote_Work']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Optional: Apply PCA to reduce dimensions for clarity
pca = PCA(n_components=2)  # Reducing to 2 components for visualization
pca_features = pca.fit_transform(scaled_features)

# Determine optimal clusters
inertia = []
silhouette_scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_features)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(pca_features, kmeans.labels_))

# Plot elbow and silhouette scores
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(2, 10), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')

plt.subplot(1, 2, 2)
plt.plot(range(2, 10), silhouette_scores, marker='o')
plt.title('Silhouette Scores')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Fit KMeans with chosen number of clusters
optimal_clusters = 4  # or the best k from elbow/silhouette analysis
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(pca_features)

# Add cluster labels to original DataFrame
sleep_remote_df['Cluster'] = clusters

# Analyze the clusters based on happiness-related features
cluster_analysis = sleep_remote_df.groupby('Cluster')[['Work_Life_Balance_Rating', 
                                                       'Satisfaction_with_Remote_Work', 
                                                       'Stress_Level_1', 
                                                       'Stress_Level_2', 
                                                       'Sleep_Quality']].agg(['mean', 'std', 'min', 'max', 'count'])
print("Cluster Analysis by Happiness Proxy Features:")
print(cluster_analysis)

# Visualize clusters
sns.scatterplot(x=pca_features[:, 0], y=pca_features[:, 1], hue=sleep_remote_df['Cluster'], palette='viridis')
plt.title("Clusters Based on Happiness-Related Features")
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.savefig('happiness_clusters.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```
